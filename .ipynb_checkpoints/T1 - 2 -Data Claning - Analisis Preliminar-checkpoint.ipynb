# Missing Values
Puede generar grandes problemas o un nivel de escala con errores altos

```python
array_null = pd.isnull(data["body"]).ravel().sum() #analizamos la columna para ver los datos .notnull(data["body"]) --> true = not null
#si agregas .values devuelve un arreglo con true and false
#si agregas .ravel() transforma en una sola fila para trabajarlo
    #con esto podemos agregarle .sum() y realizamos una sumatoria
```

```python
array_null #el valor es la cantidad de nulls que hay
```

```python
array_not_null = pd.notnull(data["body"]).ravel().sum()
array_not_null
#valores not null que hay
```

```python
#la cantidad de datos ahora si concuerda con la cantidad de tuplas
print(array_not_null+array_null)
```

#### Los valores pueden faltar por dos motivos:
* Extraccion de los datos. --> datos no nesesarios o errores de carga
* Recoleccion de datos. --> hay datos que no puedieron ser recogidos en una encuesta por x razon

hay que darles un enfoque a estos tipos de valores faltantes porque nos pueden complicar el calculo mas adelante

## Trabajo de datos
Existen dos formas de trabajar los datos faltantes, un enfoque drasticos o un enfoque no drastico.
#### Drastico:
Si no existe un dato, hay que borrar TODA la fila o columna
* la fila es conveniente cuando mucho mas de la mitad de los datos sean faltantes, pero para nada conveniente cuando solo un dato es el que falte.
* borrar la tupla es conveniente cuando existen excepciones que no afectan la logica general del esquema
vamos a usar la funcion dropna


#### No drastico:
Rellenar con nuevos valores
* vamos a utilizar la funcion fillna = la cual nos va a permitir cambiarlo de forma selectiva a traves de algun methodo
    

```python
dataSinNull = data.dropna(axis=0, how="all") 
#axis = 0 borra la fila y axis = 1 borra toda la columna
#how = bajo que condiciones borro, por ejemplo
    #all = borra si todos los valores son nulos
    #any = si alguna tiene null
dataSinNull
```

# Variables Dummy
    Variables artificiales, variables basuras.
    Se trata de un proceso que consiste en crear variables separadas para cada categoria, para el caso de tener una variable categorica.
    A pesar de que las variables categoricas ya contienen gran cantidad de informacion, a veces es util mostrar una relacion de causalidad con una sola variable de salida sin necesidad de procesarlo. Es decir cnvertir esa categoria en una nueva columna.
    El ejemplo de nuestra dataset anterior es la columna SEX.
## Que haremos?
    Convertiremos la categoria sex es dos variables dummy por ejemplo male 1 si es true or 0 si es false

```python
data["sex"]
```

```python
dummy_sex = pd.get_dummies(data["sex"],prefix="sex")
dummy_sex.head()
#Other example
#dummy_state = pd.get_dummies(data["home.dest"])
#dummy_state.head()
```

Este proceso se denomina *dummisificacion*.
Creamos una nueva tabla con dos columnas nuevas

```python
column_name = data.columns.values.tolist()
column_name
```

```python
#Borrar la columna
data = data.drop(["sex"], axis = 1)
```

```python
data
```

```python
#concatenar dos tablas
data = pd.concat([data,dummy_sex], axis=1)
```

```python
data
```

```python
#Vamos a hacer esto mas facil a traves de una funcion:
def create_dummys(df,var_name):
    dummy = pd.get_dummies(df[var_name],prefix=var_name)
    df = df.drop(var_name,axis = 1)
    df = pd.concat([df,dummy], axis=1)
    return df
```
