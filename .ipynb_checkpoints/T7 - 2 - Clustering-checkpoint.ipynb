### Metodos mas efectivos para determinar el K
Estos metodos nos ayudaran a definir el numero optimo de clusters para el dataset
* Recordar que la opinion del experto es muy importante
#### El metodo del Codo
Es una buena forma de medir la compasidad entre cluster. Lo que hacemos es medir la distancia del varicentro con respecto de los otros cluster. Obtener la suma de los cuadrado internos. Tipicamente esto se normaliza.

Estos valores son muy interesantes y nos determina la varianza.

Incrementar el numero de cluster provocara evidentemente que disminuya la suma de los cuadrados internos.

Se puede utilizar un algoritmo y determinarlo a traves de diagrams.
#### El coeficiente de la silueta
Consiste en determinar que tan cerca un punto esta de su cluster y lo alejado que esta de otros.

* Mientas menor es su valor, mas compacto es el cluster.





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
#import seaborn as sns

from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_samples , silhouette_score
```

```python
x1 = np.array([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
X = np.array(list(zip(x1,x2))).reshape(len(x1), 2)
```

```python
plt.plot()
plt.xlim([0,10])
plt.ylim([0,10])
plt.title("Dataset a clasificar")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x1,x2)
plt.show()
```

```python
max_k = 10## maximo número de clusters que vamos a crear
K = range(1,max_k)
#suma de los cuadrados internos
ssw = [] 
cmap = cm.get_cmap("Spectral") 
color_palette = [cmap(float(i)/max_k) for i in K]

centroid = [sum(X)/len(X) for i in K]
#suma de los cuadrados totales
sst = sum(np.min(cdist(X, centroid, "euclidean"), axis = 1))

#k mean por cada k
#Basicamente vamos a hacer todos lo k posibles y determinar el k optimo
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    
    centers = pd.DataFrame(kmeanModel.cluster_centers_)
    labels = kmeanModel.labels_
    
    ssw_k = sum(np.min(cdist(X, kmeanModel.cluster_centers_, "euclidean"), axis = 1))
    ssw.append(ssw_k)
    
    label_color = [color_palette[i] for i in labels]
    
    ##Fabricaremos una silueta para cada cluster
    # Por seguridad, no hacemos silueta si k = 1 o k = len(X)
    if 1<k<len(X):
       ##Crear un subplot de una fila y dos columnas
        fig, (axis1,axis2) = plt.subplots(1,2)
        fig.set_size_inches(20,8)
    
        #El primer subplot contendrá la silueta, que puede tener valores desde -1 a 1
        #En nuestro caso, ya controlamos que los valores están entre -0.1 y 1
        axis1.set_xlim([-0.1, 1.0])
        #El número de clusters a insertar determinará el tamaño de cada barra
        #El coeficiente (n_clusters+1)*10 será el espacio en blanco que dejaremos 
        #entre siluetas individuales de cada cluster para separarlas.
        axis1.set_ylim([0, len(X)+ (k+1)*10])
    
        silhouette_avg = silhouette_score(X, labels)
        print("* Para k = ",k, " el promedio de la silueta es de :",silhouette_avg)
        sample_silhouette_values = silhouette_samples(X, labels)
        
        y_lower = 10
        for i in range(k):
            #Agregamos la silueta del cluster i-ésimo
            ith_cluster_sv = sample_silhouette_values[labels == i]
            print("   - Para i = ", i+1, " la silueta del cluster vale : ", np.mean(ith_cluster_sv))
            #Ordenamos descendientemente las siluetas del cluster i-ésimo
            ith_cluster_sv.sort()
        
            #Calculamos donde colocar la primera silueta en el eje vertical
            ith_cluster_size = ith_cluster_sv.shape[0]
            y_upper = y_lower + ith_cluster_size
            
            #Elegimos el color del cluster
            color = color_palette[i]
            
            #Pintamos la silueta del cluster i-ésimo
            axis1.fill_betweenx(np.arange(y_lower, y_upper),
                               0, ith_cluster_sv, facecolor = color, alpha = 0.7)
            
            
            #Etiquetamos dicho cluster con el número en el centro
            axis1.text(-0.05, y_lower + 0.5 * ith_cluster_size, str(i+1))
            
            #Calculamos el nuevo y_lower para el siguiente cluster del gráfico
            y_lower = y_upper + 10 #dejamos vacías 10 posiciones sin muestra
            
        axis1.set_title("Representación de la silueta para k = %s"%str(k))
        axis1.set_xlabel("S(i)")
        axis1.set_ylabel("ID del Cluster")
        
        ##Fin de la representación de la silueta
        
    ##Plot de los k-means con los puntos respectivos
    plt.plot()
    plt.xlim([0,10])
    plt.ylim([0,10])
    plt.title("Clustering para k = %s"%str(k))
    plt.scatter(x1,x2, c=label_color)
    plt.scatter(centers[0], centers[1], marker = "x")
    plt.show()
```

#### Representación del codo

```python
plt.plot(K, ssw, "bx-")
plt.xlabel("k")
plt.ylabel("SSw(k)")
plt.title("La técnica del codo para encontrar el k óptimo")
plt.show()
```

#### Representación del codo normalizado


```python
plt.plot(K, 1-ssw/sst, "bx-")
plt.xlabel("k")
plt.ylabel("1-norm(SSw(k))")
plt.title("La técnica del codo normalizado para encontrar el k óptimo")
plt.show()
```

## Debilidad de KMeans
La dificultad de encontrar los K-Means

Cerca del 2007 se encontro un nuevo algoritmo para encontrar la cantidad de cluster.
Esta tecnica se donomina `propagacion de afinidad`, la metodologia utiliza las tecnicas que hemos aprendido a utilizar hasta entonces, donde analiza los puntos que se encuentran en nuestro dataset.

**Paso de mensaje** es el concepto y algoritmo clave a entender para desarrollar este algoritmo.

Se utiliza la distancia de `euclida` y mientras mas cerca de 0 es que se encuentran menos compactos y mas cerca de 1 quiere decir que estan mas proximos.

https://es.qwe.wiki/wiki/Affinity_propagation



### Implemetancion de propagacion de afinidad


```python
import pandas as pd
import numpy as np
from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
```

```python
centers = [[1,1],[1,-1],[-1,-1]]
X , labels = make_blobs(n_samples=300,centers=centers , cluster_std=0.5, random_state=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import cycle 
```

```python
sns.jointplot(x=X[:,0], y=X[:,1])
```

```python
plt.figure(figsize=(15,8))
plt.scatter(X[:,0],X[:,1],c=labels,cmap="autumn")
```

```python
#valor de preferencia de partida
af = AffinityPropagation(preference=-50).fit(X)
```

```python
#el id, de los centros de esos puntos
cluster_index = af.cluster_centers_indices_
cluster_index
```

```python
#etiquetas, a que cluster pertenece cada uno
labels = af.labels_
labels
```

```python
def report_afinnty_propagation(X):
    af = AffinityPropagation(preference=-50).fit(X)
    clust_center_ids = af.cluster_centers_indices_
    n_cluset = len(clust_center_ids)
    cluster_labels = af.labels_
    
    print('Numero estimado de Clusters ',n_cluset)
    print('Homogeniedad: %0.3f'%metrics.homogeneity_score(labels,cluster_labels))
    print('Completitud: %0.3f'%metrics.completeness_score(labels,cluster_labels))
    print('R2 cuadrado: %0.3f'%metrics.adjusted_rand_score(labels,cluster_labels))
    print('Informacion mutua ajustada: %0.3f'%metrics.adjusted_mutual_info_score(labels,cluster_labels))
    print('coeficiente de la silueta: %0.3f'%metrics.silhouette_score(X,labels,metric="sqeuclidean"))
    
    plt.figure(figsize=(16,9))
    plt.clf()
    
    colors = cycle('bgrcymkbgrcymkbgrcymkbgrcymkbgrcymk')
    
    for k, col in zip(range(n_cluset),colors):
        class_nembers = (cluster_labels == k)
        cluss_center = X[clust_center_ids[k]]
        plt.plot(X[class_nembers,0], X[class_nembers,1], col+'.')
        plt.plot(cluss_center[0], cluss_center[1],'o', markerfacecolor=col,markeredgecolor='k',markersize=14)
        for x in X[class_nembers]:
            plt.plot([cluss_center[0],x[0]],[cluss_center[1],x[1]],col)
            
    plt.title('Numero estimado de Cluster %d'%n_cluset)
    plt.show()
```

```python
report_afinnty_propagation(X)
```
