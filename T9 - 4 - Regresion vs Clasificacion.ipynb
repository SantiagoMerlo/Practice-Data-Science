# Clasificacion de Flores Iris

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn import svm, datasets
```

```python
style.use('ggplot')
plt.rcParams['figure.figsize'] = (16,8)
```

```python
iris = pd.read_csv('../python-ml-course-master/datasets/iris/iris.csv')
iris
```

```python
Species = ['setosa','virginica','versicolor']
tmp = []
for i in range(150):
    if iris['Species'][i]==Species[0]:
        tmp.append(0)
    elif iris['Species'][i]==Species[1]:
        tmp.append(1)
    else:
        tmp.append(2)
```

```python
#Analizamos solo las primeras dos porque habian sido las mas importantes en el analisis del arbol
columns = iris.columns.values.tolist()
X = iris[columns[:2]]
Y = pd.DataFrame({'Species':tmp})
```

```python
x_min,x_max = X['Sepal.Length'].min()-1 , X['Sepal.Length'].max()+1
y_min, y_max = X['Sepal.Width'].min()-1 , X['Sepal.Width'].max()+1
h = (x_max - x_min)/100
```

```python
xx , yy = np.meshgrid(np.arange(x_min , x_max , h) , np.arange(y_min , y_max , h))

#parrilla de dibujo
x_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
C = 1.0
svc = svm.SVC(kernel='linear', C=C, decision_function_shape='ovr')
svc = svc.fit(X,Y)
Ypred = svc.predict(x_plot)
Ypred = Ypred.reshape(xx.shape)
```

```python
plt.contourf(xx,yy,Ypred, cmap=plt.cm.tab10 , alpha=0.3)
plt.scatter(X['Sepal.Length'] ,X['Sepal.Width'] , c=Y['Species'] )
plt.xlabel(' Longitud del petalo ')
plt.ylabel(' Anchura del patelo ')
plt.title(' Clasificacion SVC por kernel lineal ', size=30)
```

```python
C = 1.0
svc = svm.SVC(kernel='rbf', C=C, decision_function_shape='ovr')
svc = svc.fit(X,Y)
Ypred = svc.predict(x_plot)
Ypred = Ypred.reshape(xx.shape)
```

```python
plt.contourf(xx,yy,Ypred, cmap=plt.cm.tab10 , alpha=0.3)
plt.scatter(X['Sepal.Length'] ,X['Sepal.Width'] , c=Y['Species'] )
plt.xlabel(' Longitud del petalo ')
plt.ylabel(' Anchura del patelo ')
plt.title(' Clasificacion SVC por kernel RBF ', size=30)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from sklearn.utils import shuffle
```

```python
Xtrain, Xtest , Ytrain , Ytest = train_test_split(X,Y, test_size = 0.25)
```

```python
parameters = [
    {
        'kernel':['rbf'],
        'gamma': [1e-4,1e-3,1e-2, 0.1 , 0.2 , 0.5],
        'C': [1,10,100,1000]
    },
    {
       'kernel':['linear'],
        'C': [1,10,100,1000] 
    }
]
```

```python
clf = GridSearchCV(svm.SVC(decision_function_shape='ovr') , param_grid=parameters, cv=5)
clf.fit(Xtrain,Ytrain)
```

```python
clf.best_params_
```

```python
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
params = clf.cv_results_['params']
for m, s, p in zip(means, stds, params):
    print("%0.3f (+/-%0.3f) para %r"%(m, 2*s, p))
```

```python
Y_pred = clf.predict(Xtest)
print(classification_report(Y_pred,Ytest , target_names=Species))

# Presicion Verdaderos positivos
# RECALL = VerdaderosPositivos / (VerdaderosPositivos+FalsosNegativos)
```
