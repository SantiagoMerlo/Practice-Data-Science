# Reconocimiento Facil a traves de SVM

Normalmente cuando trabajamos con este tipo de casos, debemos utilizar algoritmos mas complejos donde no se aprende en base de los pixeles sino que se analiza en base de los rasgos

```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
print(faces.target_names)
```

```python
print(faces.images.shape)
#62 pixeles de altura, y 47 de anchura
```

```python
fig , ax = plt.subplots(3,5,figsize=(16,9))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(faces.images[i] , cmap='bone')
    ax_i.set(xticks=[],yticks=[],xlabel= faces.target_names[faces.target[i]])
```

```python
#estos serian los rasgos a clasificar. vamos a analizar cada pixel y ver en que categoria entra
62*47
```

Lo que realizaremos es buscar las dimensiones mas importantes para luego solo centrarnos en esta y no sobrecargar al SVM.

Lo que queremos realizar es un analisis de imagen y luego un SVM

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
pca = PCA(n_components=150, whiten=True , random_state=42)
svc = SVC(kernel='rbf' , class_weight='balanced')
model = make_pipeline(pca,svc)
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(faces.data, faces.target, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'svc__C' : [ 0.1 , 1 , 5 , 10 , 50 ],
    'svc__gamma' : [ 0.0001 , 0.0005 , 0.001 , 0.005 , 0.01 ] 
}
grid = GridSearchCV(model,param_grid)

%time grid.fit(Xtrain, Ytrain)
```

```python
print(grid.best_params_)
```

Si nos ubiera dado un parametro mas cercano a los extranos deberiamos extender aun mas la grilla de valores para colaborar que si realmente es el valor mas optimo

```python
param_grid = {
    'svc__C' : [ 5 , 6 , 7 , 8 , 9, 10 , 11 , 12 , 13 ],
    'svc__gamma' : [ 0.0005, 0.0006 ,0.0007 , 0.0008 , 0.001 , 0.0011 , 0.0012 , 0.0013  ] 
}
grid = GridSearchCV(model,param_grid)

%time grid.fit(Xtrain, Ytrain)
```

```python
print(grid.best_params_)
```

```python
classifier = grid.best_estimator_
yfit = classifier.predict(Xtest)
```

```python
fig , ax = plt.subplots(6,7, figsize=(16,9))

for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(Xtest[i].reshape(62,47), cmap='bone')
    ax_i.set(xticks=[], yticks=[])
    ax_i.set_ylabel(faces.target_names[yfit[i]].split()[-1], color='black' if yfit[i]==Ytest[i] else 'red')
    
fig.suptitle('Prediciones de las imagenes' ,size=36)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Ytest, yfit, target_names= faces.target_names))
```

```python
# precision es la presicion de haciertos
# recall es el porentaje de falsos negativos
# support, cuantas veces salieron
from sklearn.metrics import confusion_matrix
```

```python
mat = confusion_matrix(Ytest,yfit)
```

```python
import seaborn as sns; sns.set()
```

```python
sns.heatmap(mat.T , square=True,annot=True, fmt='d' , 
    cbar=True, xticklabels=faces.target_names , yticklabels=faces.target_names)
```
