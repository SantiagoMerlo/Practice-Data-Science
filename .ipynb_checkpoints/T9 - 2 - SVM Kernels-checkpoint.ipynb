# Identificar fronteras no lineales
Existe algun conjunto de puntos donde no exista una linea para separarlos?


```python
from sklearn.datasets import make_circles , make_blobs
from sklearn.svm import SVC
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
plt.rcParams['figure.figsize'] = (10,10)
```

```python
X , Y = make_circles(n_samples=200 , noise= .1, factor= .2)
```

```python
plt.scatter(X[:,0],X[:,1], c=Y)
plt.show()
```

* La solucion que se puede plantear es hacer una nueva dimension donde los puntos amarillos estan mas alto que los de arriba, de esta forma podemos dividrlos en un plano.

```python
#vamos a transformar x en una variable con distribucion normal
# sabiendo que los puntos mas alejados valen 1 y los mas cercanos al punto vale 0
r = np.exp(-(X**2).sum(1))
```

```python
r
```

```python
from mpl_toolkits import mplot3d
```

```python
def plot_3d(elev = 30, azim = 30, X=X , Y=Y , r=r):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:,0],X[:,1] , r , c=Y)
    ax.view_init(elev=elev , azim=azim)
    
    ax.set_xlabel('x0')
    ax.set_ylabel('x1')
    ax.set_zlabel('r')
```

```python
from ipywidgets import interact, fixed
```

```python
print(interact(plot_3d, X=fixed(X), Y=fixed(Y),r=fixed(r)))
```

```python
#radial basic function
rbf = SVC(kernel='rbf', C=1E6)
rbf.fit(X,Y)
```

```python
def plt_svc(model, ax=None, plot_support=True):
    """Plot de la funcin de decisin para una clasificacin en 2D con SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    ##Generamos la parrila de puntos para evaluar el modelo
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(yy,xx)
    
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)

    
    ##Representamos las fronteras y los mrgenes del SVC
    ax.contour(X,Y,P, colors="k", levels=[-1,0,1], alpha = 0.5, linestyles=["--", "-", "--"])
    
    print(model.support_vectors_)
    
    #Puntos que actuan de suport vector machine
    if plot_support:
        ax.scatter(model.support_vectors_[:,0], 
                   model.support_vectors_[:,1], 
                   s=300, linewidth=1, facecolors = "orange");
    
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1], c=Y)
plt_svc(rbf)
```

## Ajustar los parametros de SVM
Esto va a ocurrir cuando no se tiene muy en claro los clusters.

SVM tiene algunas caracteristicas que nos permite relajar el margen, Esto se logra con el parametro C (parametro de la penalizacion)

C mientras mas grande es un valor mucho mas estrictivo, para valores mas pequenios es mas light. Se puede jugar con esto para evitar overfitting

```python
X , Y = make_blobs(n_samples=100, centers= 2 , random_state=0 , cluster_std=1.2)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
X1 , Y1 = make_blobs(n_samples=100, centers= 2 , random_state=0 , cluster_std=0.8)
plt.scatter(X1[:,0],X1[:,1],c=Y1)
```

```python
fig , ax = plt.subplots(1,2, figsize=(16,8))
fig.subplots_adjust(left = 0.05 , right=0.95 , wspace=0.1)
for ax_i , C in zip(ax,[10.0 , 0.1]):
    model = SVC(kernel='linear' , C=C)
    model.fit(X1,Y1)
    ax_i.scatter(X1[:,0],X1[:,1],c=Y)
    plt_svc(model, ax_i)
    ax_i.set_title('C es igual a '+str(C), size=15)
```

Como se puede apreciar, mientras mas ajustado sea el C nos encontramos que mas ajustado es el `pasillo`.

* En la imagen uno nos encontramos con 4 elementos que no estamos muy seguros donde pertenecen.
* En la segunda imagen nos encontramos con muchisimos mas elementos que no estan separados.
