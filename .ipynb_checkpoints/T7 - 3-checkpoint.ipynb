# Distribuciones en forma de anillo

```python
from math import pi , cos , sin , radians , sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ring(r_min = 0 , r_max = 1 , n_samples = 360 ):
    angle = np.random.uniform(0, 2*pi , n_samples)
    distance = np.random.uniform(r_min, r_max,n_samples)
    data = []
    for a,d in zip(angle,distance):
        data.append([d*cos(a),d*sin(a)])
    return np.array(data)
```

```python
data1 = ring(3,5)
data2 = ring(24,27)

data = np.concatenate([data1,data2],axis=0)
labels = np.concatenate([['Salmon' for i in range(0,len(data1))],['blue' for i in range(0,len(data2))]])
plt.figure(figsize=(9,9))
plt.scatter(data[:,0],data[:,1], c = labels , s = 10)
```

## Algoritmo con K-Mean

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(2).fit(data)
```

```python
cluster = km.predict(data)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(data[:,0],data[:,1],c=cluster,s=10)
```

## Algoritmo de los K Medoides
Consiste en elegir un conjunto de puntos desde antemano, estos seran los centros y determinaran los cluster.

* El algoritmos en ves de calcular la distancia analista que can compactos estan los puntos con los otros

```python
from pyclust import KMedoids
```

```python
kmed = KMedoids(2).fit_predict(data)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(data[:,0],data[:,1],c=kmed,s=10 , cmap='autumn')
```

## Algoritmo del Clustering Espectral
Analizar con respecto de los elementos vecinos.



```python
from sklearn.cluster import SpectralClustering
```

```python
#Es una de las tecnicas mas complejas a nivel computacional
clust = SpectralClustering(2).fit_predict(data)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(data[:,0],data[:,1], c = clust, s = 10 )
```

Que se recomienda:
    * Propagacion de afinidad si no se puede estimar la K
    * Si se puede o ya lo definimos utilizamos KMeans
    * Si se necesita analizar por algun tipo de valor medio utilizamos KMeroides
    * Si es una forma de analisis mas complejo (no son linelamente separables utilizar ClusteringSpectral)
    
* Podemos estimar la k:
    * No: Propagacin de la afinidad
    * Si: Podemos usar la distancia Eucldea:
        * Si: K-Means
        * No: Buscar valores centrales:
            * Si: K-Medoides 
            * No: Los datos son linealmente separables:
                * Si: Clustering aglomerativo
                * No: Clustering Espectral
