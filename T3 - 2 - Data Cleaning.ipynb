# Agrupacion de datos por categoria

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gender = ["Male","Female"]
social = ["Poor","Middle Class","Rich"]
```

```python
#Generacion de dataset
n = 500

gender_data = []
income_data = []

for i in range(n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(social))

# N(mu , std) --> mu + std*Z
height = 160 + 10 * np.random.randn(n)
weight = 60 + 25 * np.random.randn(n)
age = 45 + 7 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
df = pd.DataFrame(
    {
        "Gender": gender_data,
        "Social Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
df.head(10)
```

```python
## Agrupacion de datos
grupo_gender = df.groupby("Gender")
```

```python
grupo_gender.groups
```

```python
for names,groups in grupo_gender:
    print(names)
    print(groups)
```

```python
grupo_gender.get_group("Female")
```

```python
double_group = df.groupby(["Gender","Social Status"])
```

```python
double_group.groups
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Aplicar operaciones sobre datos agrupados

Los group by se les puede aplicar cualquier tipo de operacion que se realizan para dataframe

* sum() Sumatoria
* mean() Promedio
* size() Tamanio de los datos
* describe() Descripcion de todos los datos 

##### Extra:
    La funcionalidad extra que tienen los groups es el metodo aggregate, que nos va a permitir hacerle operaciones a las columnas.
*Basciamente aplicar operaciones de gropuBY*

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
group_income = double_group["Income"]
```

```python
group_income.describe()
```

```python
#En este caso cree una funcion de agregacion
# lo que hago es dividir el promedio de la desviacion estandar
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": lambda x:np.mean(x)/np.std(x)
    }
)
```

```python
#Este ejemplo es para realizar calculos a todos los elementos.
double_group.aggregate([np.sum,np.mean, lambda x:np.mean(x)/np.std(x)])
```

# Filtrado de datos

```python
double_group["Age"].filter(lambda x: x.sum()>4000)
```

## Transformacion de valores
aplico una operacion a todos los elementos del dataframe

```python
#El elemento x le resto el promedio de todos y lo divido por la desviacion estandar
zscore = lambda x: (x - x.mean())/x.std()
```

```python
z_group = double_group.transform(zscore)
z_group
```

```python
plt.hist(z_group["Age"])
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
#Si ubiera valores en blancos esto lo rellenaria
double_group.transform(fill_na_mean)
```
