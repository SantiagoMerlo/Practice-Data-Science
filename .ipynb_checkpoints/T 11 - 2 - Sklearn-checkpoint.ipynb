# Analisis de la componentes principales con SkLearn

Hay una biblioteca muy buena para hacer los graficos denominada PyPlot. Es paga sin embargo vale la pena.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../python-ml-course-master/datasets/iris/iris.csv')
```

```python
X = df.iloc[:,0:4].values
y = df.iloc[:,4].values
X_std = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
# a cuantas queremos
acp = PCA(n_components=2)
```

```python
#sklearn utiliza el caso numero C que vimos con anterioridad
Y = acp.fit_transform(X_std)
```

```python
Y
```

```python
plt.figure(figsize=(16,9))
for name in ['setosa' , 'virginica', 'versicolor' ]:
    plt.scatter(Y[y==name,0],Y[y==name,1] ,label=name ,s=50)
    plt.legend(shadow=True)
plt.show()
```
